### **Importing Python library Pandas and Python clients for accessing bioinformatics databases  (Pandas for preprocessing , Cleaning and Analysis)**

```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

### **Search for target protien from the downloaded CHEMBL Database.**





```python
# Target search for colon cancer 
target = new_client.target
target_query = target.search('her2')
targets = pd.DataFrame.from_dict(target_query)
targets
```

### **Select the target by CHEMBL ID.**

```python
selected_target = targets.target_chembl_id[1]
selected_target
```

### ***Filter the IC5O Values from the Database ***

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

### **Here we are creating a dictionary named res**

```python
df = pd.DataFrame.from_dict(res)
```

```python
df
```

### **Create a CSV file of New Preprocessed Dataframe.**


```python
df.to_csv ('her2_bioactivity_data.csv', index=False)
```

```python
df
```

### **Creating Dataframe 2 without null values for standard value and canonical smiles.** 

```python
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

### **Number of Unique values are Displayed** 






```python
len(df2.canonical_smiles.unique())
```

### **Drop Duplicate values from Canonical Smiles Notation. ***





```python
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

###**( Datapre-processing of the bioactivity data Select the 3 columns (molecule_chembl_id,canonical_smiles,standard_value) and create a new Dataframe df3 for new Data. )**

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selection]
df3
```

### **Saves dataframe to CSV file**

```python
df3.to_csv ('Her2_bioactivity_Preprocessed_Bioactivitydata.csv',index=False)
```

### **Labeling compounds as either being active, inactive or Intermediate**

### **Bioactivity data is in the IC50 unit. Compounds having values of less than 1000 nM will be considered to be active while those greater than 10,000 nM will be considered to be inactive. As for those values in between 1,000 and 10,000 nM will be referred to as intermediate.**

```python
df4 = pd.read_csv('Her2_bioactivity_Preprocessed.csv')
```

### **Create a list for bioactivity threshold using index brackets define a variable, Using If and ELSE Append the list items on the basis of the molecular weight in Nanomolar as  inactive, active and intermediate.**

```python
bioactivity_threshold = []
for i in df4.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
```

### **Define a one dimentional Array which can contain data of any type, to display the appended bioactivity threshold data create a new feature as 'Class'in it assign new variable for Bioactivity_threshold as Bioactivity_class and Concatenate df4 and bioactivity_class into  new Dataframe df5**

```python
bioactivity_class = pd.Series(bioactivity_threshold, name='class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5
```

### **Saves dataframe to CSV file**

```python
df5.to_csv('Her2_bioactivity_data_curated.csv', index=False)
```
